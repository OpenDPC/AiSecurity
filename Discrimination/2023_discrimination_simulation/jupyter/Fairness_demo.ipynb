## 定义数据划分、采样等函数 

```python
def generate_dataset(df_:pd.DataFrame) -> tuple:
    '''
    labeled数据集，unlabeled数据集划分。

    参数:
        df_ (pd.DataFrame): 原始数据集

    返回:
        labeled (pd.DataFrame): 标签集

        unlabeled (pd.DataFrame): 未标签集
    '''
    df = df_.sample(frac=1.0, random_state = 20)
    df = df.reset_index()

    labeled = df.sample(frac=0.5, random_state = 24)
    unlabeled = df[~df.index.isin(labeled.index)]     #另一半的unlabel数据集

    return labeled, unlabeled
```

```python
def resample(pro, dataset):
    '''
    把一个数据集按比例抽样

    参数:
        dataset (pd.DataFrame): 待抽样数据集
        
        pro (float): 抽样比例

    返回:
        dataset_s (pd.DataFrame): 抽样出的数据集

    '''
    dataset_s = dataset.sample(frac=pro)
    return dataset_s
```

```python
def resample_num(num, dataset, rep):
    '''
    指定个数抽样数据集

    参数:
        dataset (pd.DataFrame): 待抽样数据集
        
        num (int): 抽样个数
        
        rep (bool): 是否放回

    返回:
        dataset_s (pd.DataFrame): 抽样出的数据集

    '''
    dataset_s = dataset.sample(num, replace=rep)
    return dataset_s
```

```python
def split_4_group(dataset:pd.DataFrame, A:str) -> tuple:
    ''' 
    根据敏感属性和y值将数据集分为4组

    参数:
        dataset (pd.DataFrame): 待分数据集
        
        A (str): 敏感属性

    返回:
        以敏感属性age为例
        
        GPP (pd.DataFrame): positive and protect(Y = 1, age = 1)

        GPN (pd.DataFrame): negative and protect(Y = 0, age = 1)
        
        GUP (pd.DataFrame): positive and unprotect(Y = 1, age = 0)
        
        GUN (pd.DataFrame): negative and unprotect(Y = 0, age = 0)
    '''

#     GPP_ = dataset[dataset["age"] == 1]
    GPP_ = dataset[dataset[A] == 1]
    GPP = GPP_[GPP_["y"] == 1]
    GPN = GPP_[GPP_["y"] == 0]
    GUP_ = dataset[dataset[A] == 0]
    GUP = GUP_[GUP_["y"] == 1]
    GUN = GUP_[GUP_["y"] == 0]
    return GPP, GPN, GUP, GUN
```

```python
def Pseudo_labeling(labeled:pd.DataFrame, unlabeled:pd.DataFrame, s:float, p:float) -> pd.DataFrame:
    ''' 
    根据论文中Algorithm 1，将标记数据集拟合生成模型，并用模型给未标记数据集做预测生成伪标签

    参数:
        labeled (pd.DataFrame): 标记数据集
        
        unlabeled (pd.DataFrame): 未标记数据集

        s (float): s ∈ (0, 1)，训练集和测试集的划分比例
        
        p (float): p ∈ (0, 1)，从未标记数据集中抽样的比例

    返回:
        D_new (pd.Dataframe): 训练集和伪标签的数据集合并的D_new
    '''
    kf = KFold(n_splits=5, shuffle=False)  # 5折交叉验证，初始化KFold
    unlabeled_pre = []
    D_su = resample(p, unlabeled)     #从未标记数据集里采样得到D_su
    for train_index , test_index in kf.split(labeled):  # 调用split方法切分数据
        X_train = labeled.iloc[train_index.tolist()]
        X_test = labeled.iloc[test_index.tolist()]
        y_train = X_train['y']
        y_test = X_test['y']
        model = LogisticRegression()    #训练集拟合模型
        model.fit(X_train.drop(columns = "y"), X_train['y'])  
        y_pre = model.predict(D_su.drop(columns = "y"))
        unlabeled_pre.append(y_pre)
    unlabeled_y = np.stack(unlabeled_pre)
    pesudo_y_ = (unlabeled_y.mean(axis=0) > 0.5).astype(int)
    D_su["y"] = pesudo_y_      #伪标签
    D_new = pd.concat([X_train, D_su])     #训练集+伪标签的D_su得到D_new
    return D_new

```

```python
def fair_resample(D_new:pd.DataFrame, A:str, K:int, n_s:int, p:float) -> pd.DataFrame:
    ''' 
    根据论文中Algorithm 2，由Pseudo_labeling函数返回的D_new数据集生成fair dataset

    参数:
        D_new (pd.DataFrame): Pseudo_labeling函数返回的数据集
        
        A (str): 敏感属性
        
        K (int): 模型集成个数

        n_s (int): GPP GPN GUP GUN四个分组中最小的大小
        
        p (float): p ∈ (0, 1)，从未标记数据集中抽样的比例

    返回:
        fair_data (pd.Dataframe): 公平数据集
    '''
    D_new_gpp, D_new_gpn, D_new_gup, D_new_gun = split_4_group(D_new, A)      #D_new根据保护变量和因变量分组
    n1, n2, n3, n4= len(D_new_gpp), len(D_new_gpn), len(D_new_gup), len(D_new_gun)     #每个分组的size
    groups = [(n1, D_new_gpp), (n2, D_new_gpn), (n3, D_new_gup), (n4,D_new_gun)]     
    fair_data = []
#     print(n_s)
    for i in range(K):              #生成K个fair数据集
        fair_data_list = []
        for (n_i, g) in groups:
            if n_i >= n_s:              
                D_new_i = resample_num(n_s, g, False)
            else:
                D_new_i = resample_num(n_s, g, True)
            fair_data_list.append(D_new_i)

        fair_data_i = pd.concat(fair_data_list)
        fair_data.append(fair_data_i)
    return fair_data
```

```python
def discrimination(test_dataset:pd.DataFrame, predict:pd.DataFrame, A:str) -> float:
    ''' 
    根据论文中discrimination的定义，计算歧视水平

    参数:
        test_dataset (pd.DataFrame): 测试集
        
        predict (pd.DataFrame): 测试集的预测值
        
        A (str): 敏感属性

    返回:
        discrimination (float): 歧视水平
    '''
    N_A_1 = len(test_dataset[test_dataset[A] == 1])
    N_A_0 = len(test_dataset[test_dataset[A] == 0])
    pre_1 = test_dataset[predict == 1]
    N_Y1_A1 = len(pre_1[pre_1[A] == 1])
    N_Y1_A0 = len(pre_1[pre_1[A] == 0])
    return abs(N_Y1_A1/N_A_1 - N_Y1_A0/N_A_0)
```

```python
def uniform_sampling(dataset:pd.DataFrame, protect:str="age", label:str="y") -> pd.DataFrame:
    ''' 
    uniform sample 实现

    参数:
        dataset (pd.DataFrame): 待抽样数据集
        
        label (pd.DataFrame): 待抽样数据集的预测值
        
        protect (str): 敏感属性

    返回:
        [gpp_,gpn_,gup_,gun_] (pd.DataFrame): 抽样完成数据集
    '''
    N = len(dataset)
    S_p1 = dataset[dataset[protect] == 1] # 敏感属性 为1的数据
    S_p0 = dataset[dataset[protect] == 0] # 敏感属性 为0的数据
    S_l1 = dataset[dataset[label] == 1] # label 为1的数据
    S_l0 = dataset[dataset[label] == 0] # label 为0的数据

    GPP = S_p1[S_p1[label] == 1]
    GPN = S_p1[S_p1[label] == 0]
    GUP = S_p0[S_p0[label] == 1]
    GUN = S_p0[S_p0[label] == 0]
    gpp_ = resample_num(int(len(S_p1)* len(S_l1) / N), GPP, True)
    gpn_ = resample_num(int(len(S_p1)* len(S_l0) / N), GPN, True)
    gup_ = resample_num(int(len(S_p0)* len(S_l1) / N), GUP, True)
    gun_ = resample_num(int(len(S_p0)* len(S_l0) / N), GUN, True)
    return pd.concat([gpp_,gpn_,gup_,gun_])
```

```python
def  preferential_sampling(dataset:pd.DataFrame, protect:str="age", label:str="y") -> pd.DataFrame:
    ''' 
    preferential sample 实现

    参数:
        dataset (pd.DataFrame): 待抽样数据集
        
        label (pd.DataFrame): 待抽样数据集的预测值
        
        protect (str): 敏感属性

    返回:
        sampled (pd.DataFrame): 抽样完成数据集
    '''
    N = len(dataset)
    ranker = LogisticRegression()
    ranker.fit(dataset.drop(columns = "y"), dataset['y'])  
    y_score = ranker.predict_proba(dataset.drop(columns="y"))[:,1] ## 数据集预测为1的得分
    dataset["y_score"] = y_score
    S_p1 = dataset[dataset[protect] == 1] # 敏感属性 为1的数据
    S_p0 = dataset[dataset[protect] == 0] # 敏感属性 为0的数据
    S_l1 = dataset[dataset[label] == 1] # label 为1的数据
    S_l0 = dataset[dataset[label] == 0] # label 为0的数据

    GPP = S_p1[S_p1[label] == 1].sort_values(by=['y_score'])
    GPN = S_p1[S_p1[label] == 0].sort_values(by=['y_score'])
    GUP = S_p0[S_p0[label] == 1].sort_values(by=['y_score'])
    GUN = S_p0[S_p0[label] == 0].sort_values(by=['y_score'])
    
    ## 分别按照的得分排序 4组 
    sampled = []
    wpp = len(S_p1)* len(S_l1) / (N * len(GPP))
    wpn = len(S_p1)* len(S_l0) / (N * len(GPN)) 
    wup = len(S_p0)* len(S_l1) / (N * len(GUP)) 
    wun = len(S_p0)* len(S_l0) / (N * len(GUN))
    
    print(wpp,wpn,wup,wun)
    
    if wpp > wpn:
        sampled.append(GPN[:int(wpn) * len(GPN)]) 
        sampled.append(GPN[:int ((wpn -int(wpn)) * len(GPN))])
        sampled.append(GPP[:int(wpp * len(GPP))])
    else:
        sampled.append(GPP[:int(wpp) * len(GPP)])
        sampled.append(GPP[len(GPP) - int(wpp - int(wpp)) * len(GPP):])
        sampled.append(GPN[(len(GPN) - int(wpn) * len(GPN)):])

    if wup > wun:
        sampled.append(GUN[:int(wun) * len(GUN)]) 
        sampled.append(GUN[:int ((wun -int(wun)) * len(GUN))])
        sampled.append(GUP[:int(wup * len(GUP))])
    else:
        sampled.append(GUP[:int(wpp) * len(GUP)])
        sampled.append(GUP[len(GUP) - int(wpp - int(wpp)) * len(GUP):])
        sampled.append(GUN[(len(GUN) - int(wpn) * len(GUN)):])

    return pd.concat(sampled).drop(columns="y_score")
```

```python
def draw_picture(x:list, acc:list, dis:list, xlabel:str, save_path:str):
    ''' 
    作图函数

    参数:
        x (list): 横坐标取值
        
        acc (list): 对应x取值的准确率list
        
        dis (list): 对应x取值的歧视水平list
        
        xlabel (string): x轴标签
        
        save_path (string): 存图路径

    返回:
        无
    '''
    fig = plt.figure(figsize=(6, 4))
    #左侧坐标
    ax1 = fig.add_subplot(111)
    ax1.plot(x, acc, '*-', color='red', label='LR-Acc')
    ax1.set_ylabel('Accuracy')
    plt.xlabel(xlabel)
    ax1.legend(bbox_to_anchor=(0.2, 1.1), loc=2, borderaxespad=0)
    #右侧坐标
    ax2 = ax1.twinx()
    ax2.plot(x, dis, '*-', color='b',label='LR-Dis')
    ax2.set_ylabel('Discrimination')
    ax2.set_xlabel('Same')
    ax2.legend(bbox_to_anchor=(0.8, 1.1), loc=1, borderaxespad=0)
    plt.savefig(save_path,bbox_inches = 'tight')
    plt.show()
```

```python
def run_p_simulation(dataset:pd.DataFrame, A:str, p_list:list, K:int, N:int, model_type:str):
    ''' 
    从未标记数据集中抽样比例模拟
    输入：
        dataset: pd.DataFrame 数据集
        model_type: str 分类模型类型
        A: str 敏感属性
        p_list: list 抽样比例参数列表 
        K: int 集成模型数量
        N: int 模拟次数
    返回:
        acc: list 准确率
        dis: list 歧视水平
    ''' 
    acc = []
    dis = []
    assert model_type == "lr" or model_type == "svm", "不支持模型类型"
    

    labeled, unlabeled = generate_dataset(dataset)
    X_train, X_test, y_train, y_test=train_test_split(labeled, labeled["y"], test_size = 0.2, random_state=42)
    X_test_A = X_test

    ##计算n_s n_s为0时需要重新抽样
    gpp_train, gpn_train, gup_train, gun_train = split_4_group(X_train, A)
    gpp_test, gpn_test, gup_test, gun_test = split_4_group(X_test, A)
    gpp_unlabeled, gpn_unlabeled, gup_unlabeled, gun_unlabeled = split_4_group(unlabeled, A)

    n_s = min(len(gpp_train), len(gpn_train), len(gup_train), len(gun_train), len(gpp_test), len(gpn_test), len(gup_test), len(gun_test), 
              len(gpp_unlabeled), len(gpn_unlabeled), len(gup_unlabeled), len(gun_unlabeled))
    print(n_s,len(gpp_train), len(gpn_train), len(gup_train), len(gun_train), len(gpp_test), len(gpn_test), len(gup_test), len(gun_test), 
              len(gpp_unlabeled), len(gpn_unlabeled), len(gup_unlabeled), len(gun_unlabeled))
    for j in range(len(p_list)):
        acc_list = []
        d_list = []
        pbar = tqdm(range(N))
        for k in pbar:
            pbar.set_description(f"Run: {k+1} fit") # 设置描述
            D_new = Pseudo_labeling(labeled, unlabeled, 0.8, p_list[j])
            fair_data = fair_resample(D_new, A, K, n_s, p_list[j])
            y_pre_list =[]
            for i in range(K): 
                x_train = fair_data[i].drop(columns=[A,'y'])
                y_train = fair_data[i]['y']
                x_test = X_test_A.drop(columns=[A,'y'])
                y_test = X_test_A['y']
                if model_type == "lr":                
                    model = LogisticRegression()
                if model_type == "svm":
                    model = svm.SVC()
#                 fair_data[i] = fair_data[i].drop(columns=A)     ###去除保护标签
                model.fit(x_train, y_train)
                y_pre = model.predict(x_test)
                y_pre_list.append(y_pre)
            y_pre_ensemble =np.mean(np.stack(y_pre_list), 0)
            for i in range(len(y_pre_ensemble)):
                if y_pre_ensemble[i] > 0.5:
                    y_pre_ensemble[i] = 1
                else:
                    y_pre_ensemble[i] = 0
            acc_ = metrics.accuracy_score(y_test, y_pre_ensemble)
            d_ = discrimination(X_test_A, y_pre_ensemble, A)
            acc_list.append(acc_)
            d_list.append(d_)
        print(f"p: {p_list[j]}, acc: {np.mean(acc_list)}, dis: {np.mean(d_list)}")
        acc.append(np.mean(acc_list))
        dis.append(np.mean(d_list))
    return acc,dis
```

```python
def run_k_simulation(dataset:pd.DataFrame, A:str, p:float, k_list:list, N:int, model_type:str):
    ''' 
    模型集成次数K模拟
    输入：
        dataset: pd.DataFrame 数据集
        model_type: str 分类模型类型
        A: str 敏感属性
        p: float 从未标记数据集中抽样的比例 
        k_list: list 模型集成数量参数列表
        N: int 模拟次数
    返回:
        acc: list 准确率
        dis: list 歧视水平
    ''' 
    acc = []
    dis = []
    assert model_type == "lr" or model_type == "svm", "不支持模型类型"
    labeled, unlabeled = generate_dataset(dataset)
    X_train, X_test, y_train, y_test=train_test_split(labeled, labeled["y"], test_size = 0.2, random_state=1)
    X_test_A = X_test
    ##计算n_s n_s为0时需要重新抽样
    gpp_train, gpn_train, gup_train, gun_train = split_4_group(X_train, A)
    gpp_test, gpn_test, gup_test, gun_test = split_4_group(X_test, A)
    gpp_unlabeled, gpn_unlabeled, gup_unlabeled, gun_unlabeled = split_4_group(unlabeled, A)

    n_s = min(len(gpp_train), len(gpn_train), len(gup_train), len(gun_train), len(gpp_test), len(gpn_test), len(gup_test), len(gun_test), 
              len(gpp_unlabeled), len(gpn_unlabeled), len(gup_unlabeled), len(gun_unlabeled))
    
    for j in range(len(k_list)):
        K = k_list[j]
        acc_list = []
        d_list = []
        pbar = tqdm(range(N))
        for k in pbar:
            pbar.set_description(f"Run: {k+1} fit") # 设置描述
            D_new = Pseudo_labeling(labeled, unlabeled, 0.8, p)
            fair_data = fair_resample(D_new, A, K, n_s, p)
            y_pre_list =[]
            for i in range(K): 
                x_train = fair_data[i].drop(columns=[A,'y'])
                y_train = fair_data[i]['y']
                x_test = X_test_A.drop(columns=[A,'y'])
                y_test = X_test_A['y']
                if model_type == "lr":                
                    model = LogisticRegression()
                if model_type == "svm":
                    model = svm.SVC()
#                 fair_data[i] = fair_data[i].drop(columns=A)     ###去除保护标签
                model.fit(x_train, y_train)
                y_pre = model.predict(x_test)
                y_pre_list.append(y_pre)
            y_pre_ensemble =np.mean(np.stack(y_pre_list), 0)
            for i in range(len(y_pre_ensemble)):
                if y_pre_ensemble[i] > 0.5:
                    y_pre_ensemble[i] = 1
                else:
                    y_pre_ensemble[i] = 0
            acc_ = metrics.accuracy_score(y_test, y_pre_ensemble)
            d_ = discrimination(X_test_A, y_pre_ensemble, A)
            acc_list.append(acc_)
            d_list.append(d_)
        print(f"k: {K}, acc: {np.mean(acc_list)}, dis: {np.mean(d_list)}")
        acc.append(np.mean(acc_list))
        dis.append(np.mean(d_list))
    return acc,dis
```

```python
def run_ns_simulation(dataset:pd.DataFrame, model_type:str, A:str, p:float, K:int, N:int, ns_list:list):
    ''' 
    样本量模拟
    输入：
        dataset: pd.DataFrame 数据集
        model_type: str 分类模型类型
        A: str 敏感属性
        p: float 
        K: int 集成模型数量
        N: int 模拟次数
        ns_list: list 抽样数列表
    返回:
        acc: list 准确率
        dis: list 歧视指标
    ''' 
    acc = []
    dis = []
    assert model_type == "lr" or model_type == "svm", "不支持模型类型"
    labeled, unlabeled = generate_dataset(dataset)
    X_train, X_test, y_train, y_test=train_test_split(labeled, labeled["y"], test_size = 0.2, random_state=1)
    X_test_A = X_test
    ##计算n_s n_s为0时需要重新抽样
    gpp_train, gpn_train, gup_train, gun_train = split_4_group(X_train, A)
    gpp_test, gpn_test, gup_test, gun_test = split_4_group(X_test, A)
    gpp_unlabeled, gpn_unlabeled, gup_unlabeled, gun_unlabeled = split_4_group(unlabeled, A)

    
    for j in range(len(ns_list)):
        n_s = ns_list[j]
        acc_list = []
        d_list = []
        pbar = tqdm(range(N))
        for k in pbar:
            pbar.set_description(f"Run: {k+1} fit") # 设置描述
            D_new = Pseudo_labeling(labeled, unlabeled, 0.8, p)
            fair_data = fair_resample(D_new, A, K, n_s, p)
            y_pre_list =[]
            for i in range(K): 
                x_train = fair_data[i].drop(columns=[A,'y'])
                y_train = fair_data[i]['y']
                x_test = X_test_A.drop(columns=[A,'y'])
                y_test = X_test_A['y']
                if model_type == "lr":                
                    model = LogisticRegression()
                if model_type == "svm":
                    model = svm.SVC()
#                 fair_data[i] = fair_data[i].drop(columns=A)     ###去除保护标签
                model.fit(x_train, y_train)
                y_pre = model.predict(x_test)
                y_pre_list.append(y_pre)
            y_pre_ensemble =np.mean(np.stack(y_pre_list), 0)
            for i in range(len(y_pre_ensemble)):
                if y_pre_ensemble[i] > 0.5:
                    y_pre_ensemble[i] = 1
                else:
                    y_pre_ensemble[i] = 0
            acc_ = metrics.accuracy_score(y_test, y_pre_ensemble)
            d_ = discrimination(X_test_A, y_pre_ensemble, A)
            acc_list.append(acc_)
            d_list.append(d_)
            
        print(f" ns: {n_s}, acc: {np.mean(acc_list)}, dis: {np.mean(d_list)}")
        acc.append(np.mean(acc_list))
        dis.append(np.mean(d_list))
        
    return acc,dis
```
